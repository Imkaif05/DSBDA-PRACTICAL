```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Heart.csv")
```

```python
df.head()
```

```python
df=df.drop(['Unnamed: 0'],axis=1)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.info();
```

```python
df.describe()
```

```python
kk=pd.get_dummies(df['ChestPain'],prefix='chestpain')
```

```python
kk
```

```python
zz=pd.get_dummies(df['Thal'],prefix='Thal')
zz
```

```python
oo=pd.get_dummies(df['AHD'],prefix='AHD')
oo
```

```python
df
```

```python
dff=pd.concat([df,zz,kk,oo],axis=1)
```

```python
dff
```

```python
dff.isnull().sum()
```

```python
dff.dropna(inplace=True)
```

```python
dff.isnull().sum()
```

```python
x=dff.drop(['AHD_Yes','AHD_No','AHD','ChestPain','Thal'],axis=1)
```

```python
x
```

```python
y=dff['AHD_Yes']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtr,xtest,ytr,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
xtr.shape
```

```python
xtest.shape
```

```python
ytr.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
xtr=sc.fit_transform(xtr)
```

```python
xtr.shape
```

```python
xtest=sc.transform(xtest)
```

```python
xtest.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log=LogisticRegression()
```

```python
log.fit(xtr,ytr)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("train score :{:.4f}".format(log.score(xtr,ytr)))
```

```python
y_pred=log.predict(xtest)
```

```python
print("train score :{:.4f}".format(log.score(xtest,ytest)))
```

```python
sns.regplot(x=xtest[:,1],y=y_pred,scatter_kws=None)
plt.scatter(xtest[:,1],ytest,marker='+');
```
